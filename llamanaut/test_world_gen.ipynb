```python
import numpy as np
from scipy.spatial import Voronoi, distance

def generate_regions(cells, num_regions):
    # Get the shape of the 3D array
    shape = cells.shape

    # Generate num_regions random 3D points within the shape of the array
    points = np.random.rand(num_regions, 3) * shape

    # Compute the Voronoi diagram of the points
    vor = Voronoi(points)

    # For each cell in the array, assign it the region of the nearest point
    for i in range(shape[0]):
        for j in range(shape[1]):
            for k in range(shape[2]):
                # Compute the distance to each point
                distances = distance.cdist([[i, j, k]], points)
                # Assign the cell to the region of the nearest point
                cells[i, j, k] = np.argmin(distances)

    return cells
```

```python
cells = np.zeros((512,512,32))
cells = generate_regions(cells, 500)
print(np.unique(cells, return_counts=True))
# print(cells[400:420,400:420,10:20])
```
